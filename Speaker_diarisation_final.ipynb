## Transcription

```python
# load OpenAI Whisper automatic speech transcription
import whisper

# choose among "tiny", "base", "small", "medium", "large"
# go to https://github.com/openai/whisper/ to see other
model = whisper.load_model("small")
```

```python
import tempfile
import soundfile as sf
import numpy as np
from pyannote.core import Segment
from pyannote.audio import Audio

# Get true audio duration
with sf.SoundFile("Sticks.wav") as f:
    audio_duration = len(f) / f.samplerate

audio_file = "Sticks.wav"
first_minute = Segment(0, 60)
audio = Audio(sample_rate=16000, mono=True)

for segment, _, speaker in who_speaks_when.crop(first_minute).itertracks(yield_label=True):
    # Clamp segment to file length
    start = segment.start
    end = min(segment.end, audio_duration)
    if end <= start:
        continue  # skip invalid segment

    waveform, sample_rate = audio.crop(audio_file, Segment(start, end))

    waveform_np = waveform.numpy().astype(np.float32)
    if waveform_np.ndim == 2 and waveform_np.shape[0] == 2:
        waveform_np = waveform_np.mean(axis=0, keepdims=True)
    waveform_np = waveform_np.T  # (samples, 1)

    with tempfile.NamedTemporaryFile(suffix=".wav", delete=False) as tmp:
        sf.write(tmp.name, waveform_np, sample_rate, format='WAV', subtype='PCM_16')
        result = model.transcribe(tmp.name)
        print(f"{start:06.1f}s {end:06.1f}s {speaker}: {result['text']}")
```
